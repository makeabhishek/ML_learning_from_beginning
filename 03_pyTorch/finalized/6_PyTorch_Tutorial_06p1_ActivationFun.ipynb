# Activation Functions
- Activation functions apply a non-linear transfromation to a layer output and decide wether a neuron is activatied or not.
- If there is no activation function, there will be only linear transformation between each layer. So the whole network is basically a __Linear Regression__ model, whcih is not suited for complex task.
- With non-linear activation the model perform better.

## 0. Binary Step Function
- Not used in Practise
- It is commonly used in **binary classification** and **threshold-based decision-making** in neural networks.
$$
f(x) =
\begin{cases} 
    1, & x \geq 0 \\
    0, & x < 0
\end{cases}
$$


## 1. Linear Activation
$$
f(x) = x
$$
- No transformation applied.
- Used in regression problems.

---

## 2. Sigmoid Activation
$$
f(x) = \frac{1}{1 + e^{-x}}
$$
- Outputs values between **0 and 1**.
- Commonly used in binary classification problems.
- Typically use in the last layer of a binary classification problem
  
---

## 3. Hyperbolic Tangent (Tanh)
$$
f(x) = \frac{e^x - e^{-x}}{e^x + e^{-x}}
$$
- Basically a scaled and shifted Sigmoid fucntion
- Outputs values between **-1 and 1**.
- Zero-centered, making it useful for deep networks.
- Good choise in Hidden layers

---

## 4. Rectified Linear Unit (ReLU)
$$
f(x) =
\begin{cases} 
    x, & \text{if } x > 0 \\
    0, & \text{if } x \leq 0
\end{cases}
$$
- Most commonly used activation in deep learning.
- Efficient and prevents vanishing gradients.
- If you don't know what to use, just use ReLU for hidden layers

---

## 5. Leaky ReLU
$$
f(x) =
\begin{cases} 
    x, & \text{if } x > 0 \\
    \alpha x, & \text{if } x \leq 0
\end{cases}
$$
- Allows small gradients for **negative** values of \( x \).
- Helps avoid dying ReLU problem.
- Improved version of ReLu. Tries to solve the vanishing gradient problems. It multiply the input with a with a small value for negative numbers. Multiplication factor is a very small value.
- In ReLu, starting we have 0 values, it means the gradient is also zero and the weight in the network will never updated. So these neurons will not learn anything or these neurpons are dead. So we multiply a small factor.
- When you see that the weights in your training are not updateing you can try Leaky ReLU instead of ReLU

---

## 6. Exponential Linear Unit (ELU)
$$
f(x) =
\begin{cases} 
    x, & \text{if } x > 0 \\
    e^x - 1, & \text{if } x \leq 0
\end{cases}
$$
- Similar to Leaky ReLU but smoother at \( x = 0 \).

---

## 7. Softmax Activation (âœ…)
$$
f(x_i) = \frac{e^{x_i}}{\sum_{j} e^{x_j}}
$$
- Converts logits into **probabilities**.
- Commonly used in **multi-class classification**.
- Good in lst layer in multi class classification problems. 

### Logits in PyTorch
In PyTorch, logits refer to the raw, unnormalized scores output by a neural network before applying an activation function like softmax or sigmoid.

- Why Are Logits Important?
Logits are used because:

- They provide more numerical stability when passed to loss functions like CrossEntropyLoss.
- The activation function (like softmax) is often combined inside the loss function for efficiency.
- It allows flexibility in applying different activation functions depending on the task.
- To convert logits into probabilities, we commonly apply __softmax_:

---

## 8. Gaussian Error Linear Unit (GELU)
$$
f(x) = x \cdot \Phi(x)
$$
where \( \Phi(x) \) is the cumulative distribution function (CDF) of a Gaussian distribution.

---

## 9. Approximate GELU
$$
f(x) = 0.5 x \left(1 + \tanh\left(\sqrt{\frac{2}{\pi}} \left( x + 0.044715 x^3 \right) \right)\right)
$$
- Provides a computationally efficient approximation of GELU.

---

### **Summary**
- **ReLU** is widely used in deep learning due to its simplicity.
- **Sigmoid** and **Tanh** are used for binary classification but suffer from vanishing gradients.
- **Softmax** is used for multi-class classification.
- **GELU** and **ELU** improve on ReLU for deep networks.

````python
# plot_activation_functions.py
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Define the Step Activation Function
def step_function(x):
    return np.where(x >= 0, 1, 0)

def linear(x):
    """ Linear activation function (also called "no activation").
    Args:
        x: float tensor to perform activation.
    Returns:
        x
    """
    return x


def sigmoid(x):
    """ Sigmoid activation function return a value which lies between 0 and 1.
    Args:
         x: float tensor to perform activation.
    Returns:
        1/(1+exp(-x))
    """
    return 1 / (1 + np.exp(-x))


def tanh(x):
    """ Hyperbolic tangent function (Tanh) returns a value in between -1 and 1.
    Args:
        x: float tensor to perform activation.
    Returns:
        '(e^x - e^-x) / (e^x + e^-x)'
    """
    return (np.exp(x) - np.exp(-x)) / (np.exp(x) + np.exp(-x))


def relu(x):
    """ Rectifier Linear Unit (ReLU).
    Args:
        x: float tensor to perform activation.
    Returns:
        'x' if the value of x is greater than 0 otherwise it returns 0."""
    return np.maximum(0, x)


def leaky_relu(x):
    """ Leaky Rectifier Linear Unit activation function.
    Args:
        x: float tensor to perform activation.
    Returns:
        'x' if x greater than zero, otherwise it returns a*x.
        where a is a small constant number.
    """
    return np.where(x > 0, x, x * 0.05)


def elu(x):
    """ Exponential Linear Units (ELUs)
    Args:
        x: float tensor to perform activation.
    Returns:
        'x' for positive x, and exp(x) - 1 for negative value of x.
    """
    return np.where(x > 0, x, np.exp(x) - 1)


def softmax(x):
    """ Compute softmax values for each sets of scores in x.
    Args:
        x: float tensor to perform activation.
    Returns:
        'x' with the softmax activation applied
    """
    return np.exp(x) / np.sum(np.exp(x), axis=0)


def gelu(x):
    """ Gaussian Error Linear Unit (GELU).
    Args:
        x: float tensor to perform activation.
    Returns:
        'x' with the GELU activation applied.
    """
    return x * norm.cdf(x)


def gelu_approx(x):
    """ Approximation of Gaussian Error Linear Unit (GELU).
    Args:
        x: float tensor to perform activation.
    Returns:
        'x' with the GELU activation applied.
    """
    return 0.5 * x * (1 + tanh(np.sqrt(2 / np.pi) * (x + 0.044715 * x ** 3)))
````

````python
""" Plots different types of activation function. """
x = np.linspace(-6, 6)
plt.style.use('seaborn')
fig = plt.figure(figsize=(10, 7))

# For Linear activation function
ax1 = plt.subplot(3, 3, 1, frameon=True)
ax1.plot(x, linear(x))
ax1.set_title("Linear function")

# For Sigmoid
ax2 = plt.subplot(3, 3, 2, sharex=ax1, facecolor='orange')
ax2.plot(x, sigmoid(x), '-r')
ax2.set_title("Sigmoid function")

# For hyperbolic tangent (tanh)
ax3 = plt.subplot(3, 3, 3, sharex=ax1)
ax3.plot(x, tanh(x), '-g')
ax3.set_title("Tanh function")

# For ReLU
ax4 = plt.subplot(3, 3, 4, facecolor='orchid')
ax4.plot(x, relu(x), '-b')
ax4.set_title("ReLU function")

# For leaky ReLU
ax5 = plt.subplot(3, 3, 5)
ax5.plot(x, leaky_relu(x), '-c')
ax5.set_title("Leaky ReLU function")

# For ELU
ax6 = plt.subplot(3, 3, 6, facecolor='green')
ax6.plot(x, elu(x), '-m')
ax6.set_title("ELU function")

# For softmax activation function
ax7 = plt.subplot(3, 3, 7)
ax7.plot(x, softmax(x), '-y')
ax7.set_title("Softmax function")

# For GELU activation function
ax8 = plt.subplot(3, 3, 8, facecolor='blue')
ax8.plot(x, gelu(x), '-r')
ax8.set_title("GELU function")

# For GELU approximate, same to GELU
ax9 = plt.subplot(3, 3, 9)
ax9.plot(x, gelu_approx(x), '-k')
ax9.set_title("GELU approximate function")
fig.tight_layout()

# plt.savefig("Activation functions.jpeg")
plt.show()


# Generate input values
x = np.linspace(-10, 10, 400)  # Range from -10 to 10
y = step_function(x)

# Plot the Step Function
plt.figure(figsize=(6, 4))
plt.plot(x, y, label="Step Function", color="blue")
plt.axhline(0, color='black', linewidth=0.5, linestyle='--')  # X-axis
plt.axvline(0, color='black', linewidth=0.5, linestyle='--')  # Y-axis
plt.scatter(0, 1, color='blue', edgecolors='black', s=100)  # Mark discontinuity
plt.scatter(0, 0, color='blue', edgecolors='black', s=100, facecolors='white')  # Hollow circle at x=0
plt.title("Step Activation Function")
plt.xlabel("Input (x)")
plt.ylabel("Output (f(x))")
plt.legend()
plt.grid(True)
plt.show()
````

## Logistic/Softmax regression and Cross Entropy Loss with PyTorch



## Softmax Activation
$$
f(x_i) = \frac{e^{x_i}}{\sum_{j} e^{x_j}}
$$
- Converts logits into **probabilities**.
- Commonly used in **multi-class classification**.
It applies the exponential function to each element and normalizes it by the sunm of these exponentials. So basically it squeezwe the output between 0 and 1 so we get propbabilities.
<center>
<img src='./images/softmax_layer.PNG' width=350px>
</center>
```
####################################################################
#        -> 2.0              -> 0.65                               #
# Linear -> 1.0  -> Softmax  -> 0.25   -> CrossEntropy(y, y_hat)   #
#        -> 0.1              -> 0.1                                #
#                                                                  #
#     scores(logits)      probabilities                            #
#                           sum = 1.0                              #
####################################################################

# Softmax applies the exponential function to each element, and normalizes
# by dividing by the sum of all these exponentials
# -> squashes the output to be between 0 and 1 = probability
# sum of all probabilities is 1
```

````python
import numpy as np

def softmax(x):
    return np.exp(x)/np.sum(np.exp(x), axis =0)

x = np.array([2.0, 1.0, 0.1])
outputs = softmax(x)
print('softmax numpy:', outputs)

# We can see that the highest value has the higest probability. The sum of prabablities = 1.
````

````python
import torch 
import torch.nn as nn

x = torch.tensor([2.0, 1.0, 0.1])
outputs = torch.softmax(x, dim=0) # computes along first axis as its a tensor
print('softmax numpy:', outputs)
````

## Cross- Entropy Loss
- Cross-entropy loss, or log loss, measures the performance of a classification model 
- whose output is a probability value between 0 and 1. 
- -> loss increases as the predicted probability diverges from the actual label
- A lot of times softmax is combined with cross entropy loss. So this measure the performance of our __classification__ problem, whose output is the probability between 0 and 1. 
- This can be used in multiclass problems.
- Loss increases when the predicted probabilities diversges from actual

- $Y$ : one hot encoded class labels. Lets assume if we have three classes 1,2, 3. For our case the corerct class is 1 so we put 1 for the corerct class and all others 0. This is called one hot encoding. \
- $\hat{Y}$: Probabilities (Softmax). For predicted $\hat{Y}$, we apply softmax to get probabilities
$$
D(\hat{Y},Y) = -\frac{1}{N} \cdot  \sum Y_i \cdot log(\hat{Y_i})
$$

#### Good Prediction: low cross entropy loss
$Y = [1, 0, 0]$ and $\hat{Y} = [0.7, 0.2, 0.1]$  $\rightarrow$ $D(\hat{Y},Y) \quad =$ 0.35 

#### Bad Prediction: high cross entropy loss
$Y = [1, 0, 0]$ and $\hat{Y} = [0.1, 0.3 0.6]$  $\rightarrow$ $D(\hat{Y},Y) \quad =$ 2.30

````python
def cross_entropy(actual, predicted):
    EPS = 1e-15
    predicted = np.clip(predicted, EPS, 1 - EPS)
    loss = -np.sum(actual * np.log(predicted))
    return loss # / float(predicted.shape[0])  # we have not normalized here

# y must be one hot encoded
# if class 0: [1 0 0]
# if class 1: [0 1 0]
# if class 2: [0 0 1]
Y = np.array([1, 0, 0]) 
# predictions
Y_pred_good = np.array([0.7, 0.2, 0.1])
Y_pred_bad = np.array([0.1, 0.3, 0.6])
l1 = cross_entropy(Y, Y_pred_good)
l2 = cross_entropy(Y, Y_pred_bad)
print(f'Loss1 numpy: {l1:.4f}')
print(f'Loss2 numpy: {l2:.4f}')
````

## `nn.crossEntropyLosss`

Careful!

`nn.crossEntropyLosss` already applies `nn.LogSoftmax + nn.NLLLoss` (negative log likelihood loss)

We should not apply `Softmax` by ourself!

Y has class labels, not One-Hot!
Y_pred has raw scores(logits), no Softmax!

````python
import torch 
import torch.nn as nn

# CrossEntropyLoss in PyTorch (applies Softmax) nn.LogSoftmax + nn.NLLLoss. NLLLoss = negative log likelihood loss
loss = nn.CrossEntropyLoss()
# loss(input, target)

# target is of size nSamples = 1
# each element has class label: 0, 1, or 2
# Y (=target) contains class labels, not one-hot
Y = torch.tensor([0])

# input is of size nSamples x nClasses = 1 x 3
# y_pred (=input) must be raw, unnormalizes scores (logits) for each class, not softmax
Y_pred_good = torch.tensor([[2.0, 1.0, 0.1]]) # remember these are raw values. We didnot apply Softmax
Y_pred_bad = torch.tensor([[0.5, 2.0, 0.3]])

# Compute the loss
l1 = loss(Y_pred_good, Y)
l2 = loss(Y_pred_bad, Y)

print(f'PyTorch Loss1: {l1.item():.4f}') # good prdection has lower cross entropy loss
print(f'PyTorch Loss2: {l2.item():.4f}')

# get predictions
_, predictions1 = torch.max(Y_pred_good, 1)
_, predictions2 = torch.max(Y_pred_bad, 1)
print(f'Actual class: {Y.item()}, Y_pred1: {predictions1.item()}, Y_pred2: {predictions2.item()}')
````

````python
# PyTorch allows batch loss for multiple samples
# target is of size nBatch = 3
# each element has class label: 0, 1, or 2
Y = torch.tensor([2, 0, 1])  # correct labels. Means these are the correct class from the data and probabilities 
# input is of size nBatch x nClasses = 3 x 3
# Y_pred are logits (not softmax)
Y_pred_good = torch.tensor(
    [[0.1, 0.2, 3.9], # predict class 2
    [1.2, 0.1, 0.3], # predict class 0
    [0.3, 2.2, 0.2]]) # predict class 1

Y_pred_bad = torch.tensor(
    [[0.9, 0.2, 0.1],
    [0.1, 0.3, 1.5],
    [1.2, 0.2, 0.5]])

l1 = loss(Y_pred_good, Y)
l2 = loss(Y_pred_bad, Y)
print(f'Batch Loss1:  {l1.item():.4f}')
print(f'Batch Loss2: {l2.item():.4f}')

# get predictions
_, predictions1 = torch.max(Y_pred_good, 1)
_, predictions2 = torch.max(Y_pred_bad, 1)
print(f'Actual class: {Y}, Y_pred1: {predictions1}, Y_pred2: {predictions2}')
````

## Implement in Neural Network
- This is a typical neural network in a multiclass classification problem. We want to find out which animal is there.
- We have input layer some hidden layers, some activation fucntions in between,
- At end we have a linear layer with one output for each class.
- Here we have two outputs (Dog or Cat).
- At eh final end we apply Softmax Layer to get the probabilities.
- Here the problem statement is: __Which animal?__
<center>
<img src='./images/nnWithSoftmax.PNG' width=650px>
</center>

In pYtorch we must be carefull. We use `nn.CrossEntropyLoss()`. So we must not use `Softmax` at the end.

````python
# Lets define neural network for binary and multi class classification
# Multiclass problem
class NeuralNet2(nn.Module):
    def __init__(self, input_size, hidden_size, num_classes):
        super(NeuralNet2, self).__init__()
        # Define layers: here 1 linear layer, whcih takes input size and hidden size. Than we have an activation fucntion (relu)
        # Finally the last layer, gets the hidden size and output size for ach class.
        self.linear1 = nn.Linear(input_size, hidden_size) 
        self.relu = nn.ReLU()
        self.linear2 = nn.Linear(hidden_size, num_classes)  

    # Forward method, we only apply the layers. No softmax
    def forward(self, x):
        out = self.linear1(x)
        out = self.relu(out)
        out = self.linear2(out)
        # no softmax at the end
        return out

# Create the model and use crossEntropyy loss for classification
model = NeuralNet2(input_size=28*28, hidden_size=5, num_classes=3)
criterion = nn.CrossEntropyLoss()  # (automoatically applies Softmax)
````

# Binary classification
- if we have a binary classification problem with two possible outputs. We can change our layer as per the problem. 
- Here the problem statement is change: __Is it a Dog?__
- At the end we have a linear layer only one output.
- We donot use `Softmax` function, instead use `Sigmoid` function to get the probabilities.
- If the probability is higer than 0.5 than we say YES
- In PyTorch we use Binary Cross Entropy Loss `BCELoss`. So here we must implement the Sigmoid function at the end.
<center>
<img src='./images/binaryClassifiPytorch.PNG' width=650px>
</center>

````python
# Binary classification
class NeuralNet1(nn.Module):
    def __init__(self, input_size, hidden_size):
        super(NeuralNet1, self).__init__()
        # Set out layers and activation function 
        self.linear1 = nn.Linear(input_size, hidden_size) 
        self.relu = nn.ReLU()
        # Last layer has the output size 1. Always fixed in binary classification case.
        self.linear2 = nn.Linear(hidden_size, 1) 

    # Forawrd Pass
    def forward(self, x):
        out = self.linear1(x)
        out = self.relu(out)
        out = self.linear2(out)
        # Implement: sigmoid at the end
        y_pred = torch.sigmoid(out)
        return y_pred

model = NeuralNet1(input_size=28*28, hidden_size=5)
criterion = nn.BCELoss()
````

### Different eways to use activation fucntion.
- We can use activation funciton in two ways.  We can get diffeenrt activation fucntions from `torch.nn`.\
- We can also directly used from torch API. Example `torch.relu`

1. create the fucntions as nn modules
2. use activation function directly in Forward pass. In the __init__ function we only defeine linear layer.

````python
# Binary classification
# Option 1:
class NeuralNet(nn.Module):
    def __init__(self, input_size, hidden_size):
        super(NeuralNet1, self).__init__()
        # Set out layers and activation function 
        self.linear1 = nn.Linear(input_size, hidden_size) 
        self.relu = nn.ReLU()
        self.linear2 = nn.Linear(hidden_size, 1) 
        self.sigmoid == nn.Sigmoid()

    # Forawrd Pass: call all the fucntions we defiend above.
    def forward(self, x):
        out = self.linear1(x)
        out = self.relu(out)
        out = self.linear2(out)
        # Implement: sigmoid at the end
        out = self.Sigmoid()
        return out

# Option 2: (use activation function directly in Forward pass)
class NeuralNet(nn.Module):
    def __init__(self, input_size, hidden_size):
        super(NeuralNet1, self).__init__()
        self.linear1 = nn.Linear(input_size, hidden_size) 
        self.linear2 = nn.Linear(hidden_size, 1) 

    def forward(self, x):
        out = torch.relu(self.linear1(x))
        out = torch.sigmoid(self.linear2(out))
        return out

        
model = NeuralNet(input_size=28*28, hidden_size=5)
criterion = nn.BCELoss()
````
